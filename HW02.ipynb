```python
#The analyze from the ChatBot
#Original Code
#This part creates a basic Markov model that tracks:
#Word Frequency:
word_used: A dictionary that counts how often each word appears in the text.
#Next Word Prediction:
next_word: A nested dictionary where each key is a word, and the value is another dictionary that counts how many times subsequent words appear after that word.

#Extension #1: Bigram Tracking
word_used2 = defaultdict(int)
next_word2 = defaultdict(lambda: defaultdict(int))
for i,word in enumerate(words[:-2]):
    word_used2[word+' '+words[i+1]] += 1
    next_word2[word+' '+words[i+1]][words[i+2]] += 1 

#What This Extension Does:
#Bigrams:
This extension builds a model based on bigrams (two-word sequences). 
It concatenates the current word with the next word and uses this combination as a key in word_used2 to count occurrences of two-word phrases.
#Next Word Predictions:
It updates next_word2 to track how often a third word follows each bigram. 
So, now the model reflects how likely a word is to follow a specific two-word sequence.



#Extension #2: Character-Specific Markov Chain
from collections import Counter, defaultdict
characters = Counter("\n" + avatar.character.str.upper().str.replace(' ','.') + ":")
nested_dict = lambda: defaultdict(nested_dict)
word_used2C = nested_dict()
next_word2C = nested_dict()

for i,word in enumerate(words[:-2]):
    if word in characters:
        character = word
        
    if character not in word_used2C:
        word_used2C[character] = dict()
    if word + ' ' + words[i + 1] not in word_used2C[character]:
        word_used2C[character][word + ' ' + words[i + 1]] = 0
    word_used2C[character][word + ' ' + words[i + 1]] += 1
    
    if character not in next_word2C:
        next_word2C[character] = dict()
    if word + ' ' + words[i + 1] not in next_word2C[character]:
        next_word2C[character][word + ' ' + words[i + 1]] = dict()
    if words[i + 2] not in next_word2C[character][word + ' ' + words[i + 1]]:
        next_word2C[character][word + ' ' + words[i + 1]][words[i + 2]] = 0
    next_word2C[character][word + ' ' + words[i + 1]][words[i + 2]] += 1

#What This Extension Does:
#Character Identification:
It uses a dataset avatar and counts the occurrences of characters in it. 
It prepares a mapping where each character's name is transformed to uppercase, spaces are replaced with dots, and a colon is added. 
This essentially creates a unique identifier for each character.
#Nested Dictionaries for Characters:
word_used2C and next_word2C are initialized as nested dictionaries to keep track of word usage and next-word predictions specifically for each character.
#Character-Specific Word Tracking:
The code tracks the frequency of two-word sequences specific to each character. 
When a word matches a character's name, it logs the two-word sequence and updates the count in word_used2C.
#Next Word Prediction per Character:
The next-word predictions are similarly character-specific, allowing the model to maintain separate statistics for how often words follow the two-word phrases for each character.

#Overall Functionality of the Extensions
#Extension #1:
Transforms the original model to utilize bigrams, enhancing the context of word predictions and making the chatbot's responses potentially more coherent.
#Extension #2: 
Introduces character-specific modeling, enabling the chatbot to generate responses that reflect the unique speech patterns of different characters in the dataset. 
This allows for more personalized and context-aware interactions based on the character's dialogue.

#Conclusion
Both extensions significantly enhance the functionality of the original Markovian chatbot by adding layers of context—first through bigrams and then through character-specific tracking. 
This results in a more nuanced and engaging chatbot experience.
```

```python
#3: Original code first and the more complicated final extension then
#The analyze from the ChatBot

#Character Counting:
characters = Counter("\n" + avatar.character.str.upper().str.replace(' ', '.') + ":")

This line creates a Counter object from the character column of the avatar dataset. 
It converts each character to uppercase, replaces spaces with periods, and adds a newline character before each entry. 
This prepares a count of each unique character's representation.

#Nested Dictionaries:
nested_dict = lambda: defaultdict(nested_dict)
word_used2C = nested_dict()
next_word2C = nested_dict()

nested_dict is a recursive defaultdict, allowing for deeply nested structures. word_used2C and next_word2C will store word usage and transitions for each character.

#Looping Through Words:
for i, word in enumerate(words[:-2]):

This loop iterates through the words list, excluding the last two words to prevent index errors when accessing words[i + 1] and words[i + 2].

#Character Identification:
character = word if word in characters else None

This assigns the current word to character only if it exists in the characters Counter. 
If it doesn’t, character becomes None.

#Updating Word Usage:
if character:
    word_used2C[character][f"{word} {words[i + 1]}"] += 1

If a valid character is identified, it counts the occurrence of the word pair (current word and the next word) for that character.

#Updating Next Word Transitions:
next_word2C[character][f"{word} {words[i + 1]}"][words[i + 2]] += 1

This counts how often a specific next word follows the word pair (current and next word) for the identified character.

#Understanding Bigrams:
A bigram is a pair of consecutive words in a sequence. 
In your code, you are effectively creating bigrams and tracking their occurrences and transitions. 
Here’s how the bigram dependency is implemented:
   #Building Bigrams:
word_used2C[character][f"{word} {words[i + 1]}"] += 1

This line constructs a bigram using the current word (word) and the next word (words[i + 1]). 
It counts how many times this bigram appears for each character.

   #Tracking Transitions:
next_word2C[character][f"{word} {words[i + 1]}"][words[i + 2]] += 1

Here, you're tracking what follows the bigram (the next word words[i + 2]). 
This establishes a relationship between the bigram and the word that comes after it.

#Summary of Bigram Dependency
#Word Pair Counts: 
By counting occurrences of bigrams, you can analyze how often certain pairs of words appear together, which is essential for generating coherent text.
#Next Word Transitions: 
By linking bigrams to subsequent words, you create a chain that can be used for more complex sentence generation, allowing for contextually relevant responses based on character behavior or dialogue patterns.

#Summary of the extension code
The extension essentially builds a more complex Markov model that tracks not just individual word occurrences but also the sequences of words associated with each character. 
This allows for more nuanced text generation based on character-specific contexts.
```
